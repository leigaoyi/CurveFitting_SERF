```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Lorentz function
def lorentz(x, x0, gamma):
    return (1 / np.pi) * (gamma / ((x - x0)**2 + gamma**2))

# Define the x-axis values
x = np.linspace(-10, 10, 1000)

# Define the Lorentz parameters
x0 = 0
gamma = 1

# Evaluate the Lorentz function at each x value
y = lorentz(x, x0, gamma)

# Compute the cumulative distribution function (CDF) from the Lorentz function
cdf = np.cumsum(y)
cdf /= cdf[-1] # Normalize the CDF

# Plot the Lorentz curve and the CDF
fig, ax = plt.subplots()
ax.plot(x, y, label='Lorentz curve')
ax.plot(x, cdf, label='Cumulative distribution function')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Lorentz curve and CDF')
plt.show()
```
